```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = 'Human_Resources_Split_Data.xlsx'
df = pd.read_excel(file_path)

# Display basic info
print("Data Info:")
print(df.info())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix of Numerical Features")
plt.tight_layout()
plt.show()
```

```python
satisfaction_cols = ['EnvironmentSatisfaction', 'JobSatisfaction', 'WorkLifeBalance', 'RelationshipSatisfaction']
if all(col in df.columns for col in satisfaction_cols):
    satisfaction_by_education = df.groupby('EducationField')[satisfaction_cols].mean()
    satisfaction_by_education.plot(kind='bar', figsize=(10, 6))
    plt.title("Average Satisfaction Scores by Education Level")
    plt.ylabel("Average Score")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Salary', y='JobSatisfaction', hue='Attrition')
plt.title("Monthly Income vs Job Satisfaction (by Attrition)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='ManagerRating', y='SelfRating' , estimator='mean')
plt.title("manager rating vs self rating")
#plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Age', y='YearsSinceLastPromotion')
plt.title("Age vs years since last promotion")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='EducationField', y='YearsSinceLastPromotion')
plt.title("Years Since Last Promotion by Education Level")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
    
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Attrition', y='YearsWithCurrManager', estimator='mean')  # 'estimator' will calculate the mean
plt.title("Average Years with Current Manager by Attrition")
plt.tight_layout()
plt.show()
```
